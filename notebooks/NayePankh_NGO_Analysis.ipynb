**Objective**: Analyze beneficiary reach, program completion factors, and donation inflow trends for NayePankh Foundation (2021-2025).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/cleaned_ngo_data.csv')
df.head()
```

## 1. Summary Statistics of the Cleaned Dataset

```python
df.describe()
```

## 2. Program Performance Analysis

```python
program_summary = df.groupby('Program Category').agg(
    Total_Enrollments=('Beneficiary ID', 'count'),
    Avg_Attendance=('Attendance Percentage', 'mean'),
    Completion_Rate=('Completion Success', 'mean')
).reset_index()
program_summary['Completion_Rate'] = (program_summary['Completion_Rate'] * 100).round(2)
program_summary
```

## 3. Funds Allocation vs. Spend Utilization

```python
funds_summary = df.groupby('Program Category')[['Funds Allocated', 'Funds Utilized']].sum()
funds_summary['Util_Pct'] = (funds_summary['Funds Utilized'] / funds_summary['Funds Allocated'] * 100).round(2)
funds_summary
```

## 4. Student Outcomes: Livelihood Conversion

```python
employment_df = df[df['Program Category'].isin(['Skill Development', 'Digital Literacy']) & (df['Program Completion Status'] == 'Completed')]
employment_stats = employment_df.groupby('Program Category')['Employment Obtained After Training (Yes/No)'].value_counts(normalize=True).unstack() * 100
employment_stats.round(2)
```

## 5. Correlation Heatmap

```python
plt.figure(figsize=(8,6))
numeric_cols = ['Age', 'Attendance Percentage', 'Training Hours', 'Funds Allocated', 'Funds Utilized', 'Volunteer Count', 'Satisfaction Score']
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Key Metrics')
plt.show()
```
