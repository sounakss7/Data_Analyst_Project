# Simple Data Analytics Project: NayePankh Foundation
This notebook contains a simple analysis of NayePankh Foundation NGO data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('cleaned_ngo_data.csv')
df.head()
```

### 1. Summary Statistics

```python
df.describe()
```

### 2. Completion Rates by Program

```python
df.groupby('Program Category')['Completion Success'].mean() * 100
```

### 3. Correlation Heatmap

```python
plt.figure(figsize=(7,5))
sns.heatmap(df[['Age', 'Attendance Percentage', 'Funds Allocated', 'Funds Utilized', 'Satisfaction Score']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
